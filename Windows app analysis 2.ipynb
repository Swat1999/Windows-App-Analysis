## 4. Size Column Splitting and imputing

### So we can see that the size column is having both string and numeric values. Now we can split the column having size on one side and the type of that size on another side but doing that creates a problem

### You see if we do that we will remain with 'Size' as only numeric values but what does the numeric values mean. We can see 4.59 MB so now if we split it, it will remain as 4.59 but does it mean? it can be either GB or MB or KB we may never know

### So now we have to convert those numbers. Either turn all MB into GB or Turn all GB into MB. In this case we will turn MB into GB

# ..............................

### But first lets work on the Nan Values first

### In this case how do we deal with Nan Values? We cannot as per my knowledge because:
### 1. The column has both numeric and string values.
### 2. Splitting them will not make python understand if the size is of MB or GB
### 3. Making a condition where every value must be multiplied by 1000 if the Column has GB written on it won't work as well because the column has two data types. Even if we split it, that column will still have Nan values hence

### The only option remaining is to drop the rows having Nan values.

```python
daf_drop.dropna(subset = ['Size'], inplace=True)
```

```python
daf_drop.isnull().sum()
```

```python
daf_drop.dropna(subset=['Date of Release'], inplace=True)
```

```python
daf_drop.isnull().sum()
```

### So now that we have got a clean dataset, Lets work on the size column

```python
daf_drop[['Size', 'Size_type']] = daf_drop['Size'].str.split(' ', 1, expand=True)
```

```python
daf_drop['Size'] = daf_drop['Size'].astype(float)
```

```python
def multiple_size(row):
    if row["Size_type"] == 'GB':
        return row['Size'] * 1000.00
    else:
        return row['Size'] * 1.00
```

```python
daf_drop['Actual_size'] = daf_drop.apply(multiple_size, axis=1)
```

```python
daf_drop.tail(10)
```

```python
daf_drop.drop(['Size', 'Size_type'], axis=1)
```

### There we go Bingo, now we have a Size column that we can actually understand
